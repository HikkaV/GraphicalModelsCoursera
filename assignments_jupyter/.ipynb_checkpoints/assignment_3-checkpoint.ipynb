# ComputeTripletFactors.m

```octave
triplets = models.tripletList;
```

```octave
sample_cases.Part3SampleImagesInput(1).img
```

```octave
n
```

```octave
n = length(images);

% If the word has fewer than three characters, then return an empty list.
if (n < 3)
    factors = [];
    return
end

factors = repmat(struct('var', [], 'card', [], 'val', []), n - 2, 1);
card = [K K K];
val = ones(card);
for i=1:length(triplets),
    val(triplets(i).chars(1),triplets(i).chars(2),triplets(i).chars(3)) = triplets(i).factorVal;
end
for i=1:n-2,
    factors(i).var = [i i+1 i+2];
    factors(i).card = card;
    factors(i).val = val(:)./triplets(end).factorVal;
end
```

```octave
[charAcc, wordAcc] = ScoreModel(data.allWords, imageModel, models.pairwiseModel, models.tripletList)
```
